```python
import investpy
import json
```

```python
# SOJA
search_results = investpy.search_quotes(
    text='soy', 
    products=['commodities'], 
    countries=['brazil'], 
    n_results=10)
for search_result in search_results[:10]:
  soy = search_result.retrieve_historical_data(from_date='01/01/2022', to_date='19/05/2022')
  print(soy)
```

```python
soy['Close'].plot(figsize=(20,10), kind = 'line')
```
